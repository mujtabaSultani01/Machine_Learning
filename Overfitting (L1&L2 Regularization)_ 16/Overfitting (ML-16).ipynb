## Model Overfitting
**Model overfitting** occurs when a model is overly complex and has been trained using too much data, resulting in a model that does not generalize well to unseen data. This can lead to poor predictions when the model is used on new data. L1 and L2 regularization are two techniques used to prevent overfitting by adding a penalty to the cost function of the model. L1 regularization adds an absolute value penalty to the cost, while L2 adds a squared penalty. The penalty can be adjusted to balance between model complexity and accuracy.

```python
# Requied libraries...
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# To suppress warning for clean notebook:
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the dataset:
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
# Let's have some explorations. To see the shape of dataset:
df.shape
```

```python
# To see the unique values:
df.nunique()
```

```python
# # Now we discard some columns which is not useful for our case:
new_cols = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
new_df = df[new_cols]
new_df.head()
```

```python
# Now the next step is to check the n.a. values:
new_df.isna().sum()
```

* So it's telling us that 'Regionname' column has 3 n.a. values, 'Bedroom2' has 8217 n.a. values and so on...

```python
# So here we fill some columns with zero values, they are:
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
new_df[cols_to_fill_zero] = new_df[cols_to_fill_zero].fillna(0)
new_df.head()
```

```python
# Now if we check those columns (filled with zero):
new_df.isna().sum()
```

```python
# Some of the columns need to be filled with the mean() of that column, so they're:
new_df['Landsize'] = new_df['Landsize'].fillna(new_df.Landsize.mean())
new_df['BuildingArea'] = new_df['BuildingArea'].fillna(new_df.BuildingArea.mean())
new_df.head()
```

```python
# Now again to check:
new_df.isna().sum()
```

```python
# Now as we see our Dataset is big, and we can drop those rows which have n.a. values. 
# So here we drop the n.a. rows of 'Regionname', 'CouncilArea' and 'price' columns.
new_df.dropna(inplace = True)
new_df.isna().sum()
```

```python
new_df.head()
```

* So now see the Dataset is having no n.a. values, so we can use it for prediction.

```python
# Now again before giving the dataset into algorithm, we see, we have some columns which have categorical values. So we need
# to convert the categorical values into numbers. Here we use one-hot-encoding (dummy encoding). So we get dummies columns 
# using 'get_dummy' method:
new_df = pd.get_dummies(new_df, drop_first=True)
new_df.head()
```

```python
# So now we can create 'X' and 'Y':
# 'X' is:
X = new_df.drop(["Price"], axis = "columns")
X.head()
```

```python
# 'Y' is:
Y = new_df.Price
Y.head()
```

```python
# Now we'll do train_test_split for creating train and test samples:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 2)
```

```python
# Now we use Linear Regression to train the model, let's import it:
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Now let's train the model:
lr.fit(X_train, y_train)
```

```python
# The model is trained. To check the model accuracy:
lr.score(X_test, y_test)
```

```python
# We see the model scores is very low. If check the score on trainin dataset:
lr.score(X_train, y_train)
```

```python
# So on the training data the scores is lookin good. So this is clearly overfitting the dataset. It give us a good score for
# train dataset, but for test dataset which has been not seen by the model before, it gives the horible scores.
# So to address the overfitting problem, sklearn provide a model called 'lasso', lasso is simply L1 Regularization. 
# we going to use this model:

from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)

# We can play with the 'alpha' parameter, to check which one give us the good result.
# So now the model is fitting with the regularization parameters... [As L1 equation (absolut Teta value)]
```

```python
# Now to check accuracy with using lasso:
lasso_reg.score(X_test, y_test)
```

* As we see the score is getting better.

```python
# To see the score on train dataset:
lasso_reg.score(X_train, y_train)
```

* So we see training and testing both giving us good result, not good result but better than previous situation.

```python
# The next regularization technique is L2 Regularization which is called 'Ridge', let's use it:

from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
# Now to check the scores on test dataset:
ridge_reg.score(X_test, y_test)
```

```python
# To check the scores on train dataset:
ridge_reg.score(X_train, y_train)
```

* So we see again for both train and test datasets it gives better result comparing with previous result without using regularization techniques.
