## Outlier Detection & Removal using Z_score and Standard Deviation


```python
# Required libraries:
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10, 6)
```

```python
# Let's first load the dataset into the Jupyter notebook:
df = pd.read_csv("heights.csv")
df.head()
```

```python
# The first thing we want to do is to plot the data into histograms to see the data distribution (if it's normal or not?):
plt.hist(df.height, bins=20, rwidth=0.8)
plt.xlabel('Height (inches)')
plt.ylabel('Count')
plt.show()
```

* The histogram says us that for a specific height which is on 'X' axis how many samples do we have? which is plotted on 'Y' axis. For example for 65 inches height there is more than 1000 samples.
* Normal distribution means you have a shape of 'Bell Curve' in plotted histograms. The majority of values are around the means or everage. When you go away from mean the values become less. for more about normal distribution go to https://www.mathsisfun.com/data/standard-normal-distribution.html link.

```python
# So the upper dataset is clearly following the normal distribution. Let's plot the 'Bell Curve' for the histogram. For that we need a 'sciepy' model:
from scipy.stats import norm
import numpy as np

plt.hist(df.height, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(df.height.min(), df.height.max(), 0.1)         # Produce the range of 'X' values. min and max give you the minimum and maximum values for the height. To see them individually, you can print it as 'df.height.min()' and 'df.height.max()'. Or you can 'df.describe()' to tell you the quick statistics.
plt.plot(rng, norm.pdf(rng,df.height.mean(),df.height.std()))  # ...
```

```python
# To print the mean of the height column:
df.height.mean()
```

```python
# To print the standard deviation of the height column:
# Standard deviation is basically showing you how far a way the data points is from mean values? For example our mean is 66,
# so if we have 68 value, we can say that this data point is close to the mean. But if we have 88 value, then we say that
# this data point is very far away from the mean. The three types of standard deviation will be check in the mathsisfun site.
df.height.std()
```

```python
# Now what standard deviation say is: if we have a data point whihc is in 3.84 range with mean (66.36) so it's 1-standard deviation (66.36 + 3.84 is 1-standard deviation and -3.84 + 66.36 is 2-standard deviation)
# So here first we use 3-standard deviation to remove the outlier. 3-sd is a common practice in the industry to remove the outlier, sometime people use 4-sd or 5-sd to remove the outlier. if the dataset is small then it's better to use 2-sd.
# One of the ways we can remove outliers is remove any data points that are beyond 3-standard deviation from mean. Which means we can come up with following upper and lower bounds:
upper_limit = df.height.mean() + 3*df.height.std()
upper_limit
```

* Any thing that is upper then 77 will be mark as outlier.

```python
# We can do the similar thing for lower-limit:
lower_limit = df.height.mean() -3*df.height.std()
lower_limit
```

* Any thing that is lower then 54 will be mark as outlier.

```python
# Now to see what are the outlier in our DataFrame:
df[(df.height>upper_limit) | (df.height<lower_limit)]
```

* So these 7 data points are highlighted as outlier.

```python
# To remove the upper outliers:
df_no_outlier_std_dev = df[(df.height<upper_limit) & (df.height>lower_limit)]
df_no_outlier_std_dev.head()
```

```python
# To see the valid samples:
df_no_outlier_std_dev.shape
```

```python
# To show the number of rows which we removed:
df.shape[0] - df_no_outlier_std_dev.shape[0]
```

### Outlier detection and removal using Z Score

**Z score is a way to achieve same thing that we did above in part (1)**
  
**Z score indicates how many standard deviation away a data point is.**

For example in our case mean is 66.37 and standard deviation is 3.84.

If a value of a data point is 77.91 then Z score for that is 3 because it is 3 standard deviation away (77.91 = 66.37 + 3 * 3.84)

<img src = "zscore.png" width = 300 height = 300/>

```python
# Now we use Z_scores to do the same. Z_scores is a kind similar to standard deviation, Z_scores tell us that how many standard deviation you are a way from the mean.
# Let's say if your datapoint is three standard deviation a way from the mean, then the Z_score for that data point is 3.
# Let's create a new column and add the Z_scores:
df['zscore'] = ( df.height - df.height.mean() ) / df.height.std()
df.head(5)
```

```python
# To see how we get the Z_scores, for example 1.943964? 
(73.84-66.37)/3.84
```

```python
# To see all the data points which has Z_scores more than 3: 
df[df['zscore']>3]
```

```python
# Similarly to get data points having Z_scores less than 3:
df[df['zscore'] < -3]
```

```python
# To view all the outliers: 
df[(df.zscore<-3) | (df.zscore>3)]
```

```python
# To filter the outliers, we use the same apporach as prebious:
df_no_outliers = df[(df.zscore>-3) & (df.zscore<3)]
df_no_outliers.head()
```

```python
# To show the number of rows which we removed:
df.shape[0] - df_no_outliers.shape[0]
```

### Exercise
You are given bhp.csv which contains property prices in the city of banglore, India. You need to examine price_per_sqft column and do following,

1. Remove outliers using percentile technique first. Use [0.001, 0.999] for lower and upper bound percentiles

2. After removing outliers in step 1, you get a new dataframe.

3. On step(2) dataframe, use 4 standard deviation to remove outliers

4. Plot histogram for new dataframe that is generated after step (3). Also plot bell curve on same histogram

5. On step(2) dataframe, use zscore of 4 to remove outliers. This is quite similar to step (3) and you will get exact same result

```python
# Let's load the dataset:
dfe = pd.read_csv("bhp.csv")
dfe.head()
```

```python
# Let's see the shape of dataset:
dfe.shape
```

```python
# Let's have a short statistics of the dataset:
dfe.describe()
```

```python
# Let's using percentile define the upper and lower limits for the 'price_per_sqft' column:
min_threshold, max_threshold = dfe.price_per_sqft.quantile([0.001, 0.999])
min_threshold, max_threshold
```

```python
# So now let's see which data points have values outside the defined min creteria:
dfe[dfe.price_per_sqft < min_threshold]
```

```python
# Let's see which data points have value outside the defined max creteria:
dfe[dfe.price_per_sqft > max_threshold]
```

```python
# To remove the outliers:
new_dfe = dfe[(dfe.price_per_sqft<max_threshold) & (dfe.price_per_sqft>min_threshold)]
new_dfe.sample(10)
```

```python
# Let's define the upper limit for 4-sd:
upper_limit = new_dfe.price_per_sqft.mean() + 4*new_dfe.price_per_sqft.std()
upper_limit
```

```python
# We can do the similar thing for lower-limit:
lower_limit = new_dfe.price_per_sqft.mean() -4*new_dfe.price_per_sqft.std()
lower_limit
```

```python
# Now to see what are the outliers in our DataFrame:
new_dfe[(new_dfe.price_per_sqft>upper_limit) | (new_dfe.price_per_sqft<lower_limit)]
```

```python
# To remove the upper outliers:
new_dfe_no_outlier_std_dev = new_dfe[(new_dfe.price_per_sqft<upper_limit) & (new_dfe.price_per_sqft>lower_limit)]
new_dfe_no_outlier_std_dev.head()
```

```python
# To see the shape of dataset:
new_dfe_no_outlier_std_dev.shape
```

```python
# To show the number of rows which we removed:
new_dfe.shape[0] - new_dfe_no_outlier_std_dev.shape[0]
```

```python
# To plot the dataset using histogram:
plt.hist(new_dfe_no_outlier_std_dev.price_per_sqft, bins=20, rwidth=0.8)
plt.xlabel('Height (inches)')
plt.ylabel('Count')
plt.show()
```

```python
# To display the 'Bell Curve':
plt.hist(new_dfe_no_outlier_std_dev.price_per_sqft, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(new_dfe_no_outlier_std_dev.price_per_sqft.min(), new_dfe_no_outlier_std_dev.price_per_sqft.max(), 0.1)      
plt.plot(rng, norm.pdf(rng,new_dfe_no_outlier_std_dev.price_per_sqft.mean(),new_dfe_no_outlier_std_dev.price_per_sqft.std()))  
```

```python
# Let's create a new column and add the Z_scores:
new_dfe['e_zscore'] = ( new_dfe.price_per_sqft - new_dfe.price_per_sqft.mean() ) / new_dfe.price_per_sqft.std()
new_dfe.head(5)
```

```python
# To view all the outliers: 
new_dfe[(new_dfe.e_zscore<-4) | (new_dfe.e_zscore>4)]
```

```python
# To filter the outliers, we use the same apporach as prebious (standard divation):
new_dfe_no_outliers = new_dfe[(new_dfe.e_zscore>-4) & (new_dfe.e_zscore<4)]
new_dfe_no_outliers.head()
```

```python
# To show the number of rows which we removed:
new_dfe.shape[0] - new_dfe_no_outliers.shape[0]
```
